```python
#Importing Required Libraries

from tkinter import *
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB

#Reading the data

data = pd.read_csv("C://Users//anith//OneDrive//Desktop//Brain Tumor//Tumor Data.csv")

#Preprocessing the data

df={'Headache':{'Absent':0,'Severe':1,'Common':2}}
data1=data.replace(df,inplace=True)

df2={'Vision':{'Normal':0,'Double vision':1,'Blurred vision':2,'Blind':3}}
data2=data.replace(df2,inplace=True)

df3={'Vomiting':{'Absent':0,'Present':1}}
data3=data.replace(df3,inplace=True)

df4={'Hearing':{'Yes':0,'No':1}}
data4=data.replace(df4,inplace=True)

df5={'Personality':{'Normal':0,'Decreasing':1}}
data5=data.replace(df5,inplace=True)

df6={'Patience':{'Normal':0,'Decreasing':1}}
data6=data.replace(df6,inplace=True)

df7={'Behavior':{'Normal':0,'Abnormal':1}}
data7=data.replace(df7,inplace=True)

df8={'Sleep':{'Adequate':0,'Inadequate':1}}
data8=data.replace(df8,inplace=True)

df9={'Muscular':{'Normal':0,'Muscle Weakness':1,'Difficulty Walking':2}}
data9=data.replace(df9,inplace=True)

#Preparing Training and Testing Data

x=data.drop(['Class'],axis=1)
y=data['Class']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=10)

#Training Model 1

lr=LogisticRegression()
lr.fit(x_train,y_train)

#Testing Model 1

y_pred1=lr.predict(x_test)
accuracy1=accuracy_score(y_test,y_pred1)

#Training Model 2

rf=RandomForestClassifier()
rf.fit(x_train,y_train)

#Testing Model 2

y_pred2=rf.predict(x_test)
accuracy2=accuracy_score(y_test,y_pred2)

#Training Model 3

dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)

#Testing Model 3

y_pred3=dtc.predict(x_test)
accuracy3=accuracy_score(y_test,y_pred3)

#Training Model 4

nn=KNeighborsClassifier()
nn.fit(x_train,y_train)

#Testing Model 4

y_pred4=nn.predict(x_test)
accuracy4=accuracy_score(y_test,y_pred4)

#Training Model 5

svc=SVC()
svc.fit(x_train,y_train)

#Testing Model 5

y_pred5=svc.predict(x_test)
accuracy5=accuracy_score(y_test,y_pred5)

#Training Model 6

nb=GaussianNB()
nb.fit(x_train,y_train)

#Testing Model 6

y_pred6=nb.predict(x_test)
accuracy6=accuracy_score(y_test,y_pred6)

#Creating User Interface

root=Tk()
root.title("Brain Tumor Detection")
root.geometry('400x750')

label1=Label(root,text="Headache :")
label1.grid(row=0,column=0,padx=5,pady=10)

clicked1=StringVar()
clicked1.set("Select")
option1=OptionMenu(root,clicked1,"Absent","Severe","Common")
option1.grid(row=0,column=1,pady=10)

label2=Label(root,text="Vision :")
label2.grid(row=1,column=0,padx=5,pady=10)

clicked2=StringVar()
clicked2.set("Select")
option2=OptionMenu(root,clicked2,"Normal","Double Vision","Blurred Vision","Blind")
option2.grid(row=1,column=1,pady=10)

label3=Label(root,text="Vomiting :")
label3.grid(row=2,column=0,padx=5,pady=10)

clicked3=StringVar()
clicked3.set("Select")
option3=OptionMenu(root,clicked3,"Absent","Present")
option3.grid(row=2,column=1,pady=10)

label4=Label(root,text="Hearing :")
label4.grid(row=3,column=0,padx=5,pady=10)

clicked4=StringVar()
clicked4.set("Select")
option4=OptionMenu(root,clicked4,"Yes","No")
option4.grid(row=3,column=1,pady=10)

label5=Label(root,text="Personality :")
label5.grid(row=4,column=0,padx=5,pady=10)

clicked5=StringVar()
clicked5.set("Select")
option5=OptionMenu(root,clicked5,"Normal","Decreasing")
option5.grid(row=4,column=1,pady=10)

label6=Label(root,text="Patience :")
label6.grid(row=5,column=0,padx=5,pady=10)

clicked6=StringVar()
clicked6.set("Select")
option6=OptionMenu(root,clicked6,"Normal","Decreasing")
option6.grid(row=5,column=1,pady=10)

label7=Label(root,text="Behavior :")
label7.grid(row=6,column=0,padx=5,pady=10)

clicked7=StringVar()
clicked7.set("Select")
option7=OptionMenu(root,clicked7,"Normal","Abnormal")
option7.grid(row=6,column=1,pady=10)

label8=Label(root,text="Sleep :")
label8.grid(row=7,column=0,padx=5,pady=10)

clicked8=StringVar()
clicked8.set("Select")
option8=OptionMenu(root,clicked8,"Adequate","Inadequate")
option8.grid(row=7,column=1,pady=10)

label9=Label(root,text="Muscular :")
label9.grid(row=8,column=0,padx=5,pady=10)

clicked9=StringVar()
clicked9.set("Select")
option9=OptionMenu(root,clicked9,"Normal","Muscle Weakness","Difficulty Walking")
option9.grid(row=8,column=1,pady=10)

label10=Label(root,text="Model :")
label10.grid(row=9,column=0,padx=5,pady=10)

clicked10=StringVar()
clicked10.set("Select")
option10=OptionMenu(root,clicked10,"Model 1","Model 2","Model 3","Model 4","Model 5","Model 6")
option10.grid(row=9,column=1,pady=10)

def Headache(k):
    if k=="Absent":
        return 0
    elif k=="Severe":
        return 1
    else:
        return 2

def Vision(k):
    if k=="Normal":
        return 0
    elif k=="Double Vision":
        return 1
    elif k=="Blurred Vision":
        return 2
    else:
        return 3
    
def Vomiting(k):
    if k=="Present":
        return 0
    else:
        return 1
    
def Hearing(k):
    if k=="Yes":
        return 0
    else:
        return 1
    
def Personality(k):
    if k=="Normal":
        return 0
    else:
        return 1

def Patience(k):
    if k=="Normal":
        return 0
    else:
        return 1
    
def Behavior(k):
    if k=="Normal":
        return 0
    else:
        return 1
    
def Sleep(k):
    if k=="Adequate":
        return 0
    else:
        return 1
    
def Muscular(k):
    if k=="Normal":
        return 0
    elif k=="Muscle Weakness":
        return 1
    else:
        return 2

def predict():
    model_no=clicked10.get()
    lst=[Headache(clicked1.get()),Vision(clicked2.get()),Vomiting(clicked3.get()),Hearing(clicked4.get()),
         Personality(clicked5.get()),Patience(clicked6.get()),Behavior(clicked7.get()),
         Sleep(clicked8.get()),Muscular(clicked9.get())]
    
    if model_no=="Model 1":
        val=lr.predict([lst])
        value=''.join(map(str,val))
        if value=="No":
            label=Label(root,text="No Brain Tumor Detected")
            label.grid(row=11,column=1,pady=10)
        elif value=="Yes(1)":
            label=Label(root,text="Brain Tumor Detected - Stage 1")
            label.grid(row=11,column=1,pady=10)
        elif value=="Yes(2)":
            label=Label(root,text="Brain Tumor Detected - Stage 2")
            label.grid(row=11,column=1,pady=10)
        else:
            label=Label(root,text="Brain Tumor Detected - Stage 3")
            label.grid(row=11,column=1,pady=10)
    elif model_no=="Model 2":
        val=rf.predict([lst])
        value=''.join(map(str,val))
        if value=="No":
            label=Label(root,text="No Brain Tumor Detected")
            label.grid(row=11,column=1,pady=10)
        elif value=="Yes(1)":
            label=Label(root,text="Brain Tumor Detected - Stage 1")
            label.grid(row=11,column=1,pady=10)
        elif value=="Yes(2)":
            label=Label(root,text="Brain Tumor Detected - Stage 2")
            label.grid(row=11,column=1,pady=10)
        else:
            label=Label(root,text="Brain Tumor Detected - Stage 3")
            label.grid(row=11,column=1,pady=10)
    elif model_no=="Model 3":
        val=dtc.predict([lst])
        value=''.join(map(str,val))
        if value=="No":
            label=Label(root,text="No Brain Tumor Detected")
            label.grid(row=11,column=1,pady=10)
        elif value=="Yes(1)":
            label=Label(root,text="Brain Tumor Detected - Stage 1")
            label.grid(row=11,column=1,pady=10)
        elif value=="Yes(2)":
            label=Label(root,text="Brain Tumor Detected - Stage 2")
            label.grid(row=11,column=1,pady=10)
        else:
            label=Label(root,text="Brain Tumor Detected - Stage 3")
            label.grid(row=11,column=1,pady=10)
    elif model_no=="Model 4":
        val=nn.predict([lst])
        value=''.join(map(str,val))
        if value=="No":
            label=Label(root,text="No Brain Tumor Detected")
            label.grid(row=11,column=1,pady=10)
        elif value=="Yes(1)":
            label=Label(root,text="Brain Tumor Detected - Stage 1")
            label.grid(row=11,column=1,pady=10)
        elif value=="Yes(2)":
            label=Label(root,text="Brain Tumor Detected - Stage 2")
            label.grid(row=11,column=1,pady=10)
        else:
            label=Label(root,text="Brain Tumor Detected - Stage 3")
            label.grid(row=11,column=1,pady=10)
    elif model_no=="Model 5":
        val=svc.predict([lst])
        value=''.join(map(str,val))
        if value=="No":
            label=Label(root,text="No Brain Tumor Detected")
            label.grid(row=11,column=1,pady=10)
        elif value=="Yes(1)":
            label=Label(root,text="Brain Tumor Detected - Stage 1")
            label.grid(row=11,column=1,pady=10)
        elif value=="Yes(2)":
            label=Label(root,text="Brain Tumor Detected - Stage 2")
            label.grid(row=11,column=1,pady=10)
        else:
            label=Label(root,text="Brain Tumor Detected - Stage 3")
            label.grid(row=11,column=1,pady=10)
    else:
        val=nb.predict([lst])
        value=''.join(map(str,val))
        if value=="No":
            label=Label(root,text="No Brain Tumor Detected")
            label.grid(row=11,column=1,pady=10)
        elif value=="Yes(1)":
            label=Label(root,text="Brain Tumor Detected - Stage 1")
            label.grid(row=11,column=1,pady=10)
        elif value=="Yes(2)":
            label=Label(root,text="Brain Tumor Detected - Stage 2")
            label.grid(row=11,column=1,pady=10)
        else:
            label=Label(root,text="Brain Tumor Detected - Stage 3")
            label.grid(row=11,column=1,pady=10)
    
button=Button(root,text="Predict",command=predict)
button.grid(row=10,column=1,pady=10)

def Accuracy():
    val=clicked10.get()
    if val=="Model 1":
        label11=Label(root,text="Accuracy of Model 1 is "+str(int(accuracy1*100*100)/100)+"%")
        label11.grid(row=13,column=1,pady=10)
    elif val=="Model 2":
        label11=Label(root,text="Accuracy of Model 2 is "+str(int(accuracy2*100*100)/100)+"%")
        label11.grid(row=13,column=1,pady=10)
    elif val=="Model 3":
        label11=Label(root,text="Accuracy of Model 3 is "+str(int(accuracy3*100*100)/100)+"%")
        label11.grid(row=13,column=1,pady=10)
    elif val=="Model 4":
        label11=Label(root,text="Accuracy of Model 4 is "+str(int(accuracy4*100*100)/100)+"%")
        label11.grid(row=13,column=1,pady=10)
    elif val=="Model 5":
        label11=Label(root,text="Accuracy of Model 5 is "+str(int(accuracy5*100*100)/100)+"%")
        label11.grid(row=13,column=1,pady=10)
    else:
        label11=Label(root,text="Accuracy of Model 6 is "+str(int(accuracy6*100*100)/100)+"%")
        label11.grid(row=13,column=1,pady=10)
        
button2=Button(root,text="Accuracy",command=Accuracy)
button2.grid(row=12,column=1,pady=10)

root.mainloop()
```
